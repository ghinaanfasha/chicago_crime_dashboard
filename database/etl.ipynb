```python
import pandas as pd

df = pd.read_csv("chicago_crime_500k.csv")
df.info()
```

```python
df = df.drop(['unique_key', 'case_number', 'block', 'x_coordinate', 'y_coordinate', 'year', 'updated_on', 'location'], axis=1)
df.info()
```

```python
print("data kosong : ", df.isnull().sum())
print("\ndata duplikat : ", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print("\ndata duplikat : ", df.duplicated().sum())
```

```python
df = df.dropna()
print("data kosong : ", df.isnull().sum())
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])

df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['hour'] = df['date'].dt.hour
df['weekday'] = df['date'].dt.day_name(locale='id_ID')
df.info()
```

```python
df = df.drop('date', axis=1)
df.info()
```

```python
def crime_scene(hour):
    if 1 <= hour < 5:
        return 'Dini hari'
    elif 5 <= hour < 11:
        return 'Pagi'
    elif 11 <= hour < 15:
        return 'Siang'
    elif 15 <= hour < 18:
        return 'Sore'
    else:
        return 'Malam'

df['daytime'] = df['hour'].apply(crime_scene)
df.info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['primary_type'])

kmeans = KMeans(n_clusters=5, random_state=42)
df['crime_category'] = kmeans.fit_predict(X)

X_loc = vectorizer.fit_transform(df['location_description'].astype(str))
kmeans_loc = KMeans(n_clusters=5, random_state=42)
df['location_category'] = kmeans_loc.fit_predict(X_loc)

df.head()
```

```python
df_subset = df[['iucr', 'primary_type', 'description', 'location_description', 'arrest',
       'domestic', 'beat', 'district', 'ward', 'community_area', 'fbi_code',
       'latitude', 'longitude', 'day', 'month', 'year', 'weekday', 'daytime']]
incident_counts = df_subset.groupby(['iucr', 'primary_type', 'description', 'location_description', 'arrest',
       'domestic', 'beat', 'district', 'ward', 'community_area', 'fbi_code',
       'latitude', 'longitude', 'day', 'month', 'year', 'weekday', 'daytime']).size().reset_index(name='incident_count')

incident_counts.head()
```

```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('postgresql+psycopg2://postgres:behelhijau@localhost:5432/crimedb')

dim_crime = pd.read_sql("SELECT id_crime, primary_type, description FROM dim_crime", engine)
dim_date = pd.read_sql("SELECT id_date, day, month, year, daytime, weekday FROM dim_date", engine)
dim_location = pd.read_sql("SELECT id_location, location_description FROM dim_location", engine)
dim_adm = pd.read_sql("SELECT id_adm, ward, beat, district, community_area FROM dim_area_adm", engine)
dim_class = pd.read_sql("SELECT id_class, iucr, fbi_code FROM dim_classification", engine)
```

```python
incident_counts = incident_counts.merge(dim_crime, on=['primary_type', 'description'], how='left')

incident_counts = incident_counts.merge(dim_date, on=['day', 'month', 'year', 'weekday', 'daytime'], how='left')

incident_counts = incident_counts.merge(dim_location, on=['location_description'], how='left')

incident_counts = incident_counts.merge(dim_adm, on=['ward', 'beat', 'district', 'community_area'], how='left')

incident_counts = incident_counts.merge(dim_class, on=['iucr', 'fbi_code'], how='left')
```

```python
incident_counts.head()
```

```python
from psycopg2.extras import execute_values

conn = psycopg2.connect(
    dbname="crimedb",
    user="postgres",
    password="behelhijau",
    host="localhost",
    port="5432"
)
cursor = conn.cursor()

data_to_insert = incident_counts[['id_crime', 'id_date', 'id_location', 'id_adm', 'id_class', 'latitude', 'longitude', 'domestic', 'arrest', 'incident_count']].values.tolist()

query = """
    INSERT INTO crime_facts (id_crime, id_date, id_location, id_adm, id_class, latitude, longitude, domestic_status, arrest_status, incident_count)
    VALUES %s
"""

execute_values(cursor, query, data_to_insert)

conn.commit()
cursor.close()
conn.close()

print("Data berhasil diinsert ke tabel fakta!")
```
